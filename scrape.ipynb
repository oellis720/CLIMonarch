```python
import json
from bs4 import BeautifulSoup
import urllib.request
import pandas as pd

# Pulls the url link and allows the html code to be viewed
url = "https://en.wikipedia.org/wiki/Swedish_royal_family"
page = urllib.request.urlopen(url)

# Shows the html
soup = BeautifulSoup(page, 'lxml')
#print(soup.prettify())

# Shows the html tags 'tbody' and everything that resides in it.
sort_table = soup.find_all('table')
#print(sort_table)
monarchs = sort_table[1]
#print(monarchs)
```

```python
name = []

for rows in monarchs.findAll('td', colspan='6'):
    #print(rows)
    if len(rows) == 1:
        name.append(rows.find(text=True))
    if len(rows) == 2:
        name.append(rows.find(text=True))
        
print(name)
```
